# Matriz de incidencia

## 1- Definir el stoplist

```python
with open("stoplist.txt") as file:
    stoplist = [line.strip() for line in file]
stoplist += ['?', '-', '.', ':', ',', '', '!', '', ';']    
```

## 2-  Tokenizar y eliminar stopwords

```python
import string
def preprocess(doc):
    with open(doc) as file:
        text = file.read()
    words = nltk.word_tokenize(text.lower(), language='spanish')
    relevantWords = []
    for word in words:
        if word not in stoplist:
            relevantWords.append(word)
    return relevantWords
```

### Prueba

```python
print(preprocess("docs/libro1.txt"))
```

## 3- Construir la matriz de incidencia

```python
def buildMatrix(docs):
    docIndex = {}
    i = 0
    for doc in docs:
        docIndex[doc] = i
        i  = i + 1

    wordIndex = {}
    i = 0
    for doc in docs:
        words = preprocess(doc)
        for word in words:
            if word not in wordIndex.keys():
                wordIndex[word] = i
                i = i + 1

    G = np.zeros((len(wordIndex), len(docIndex)), dtype=bool)

    for doc in docs:
        words = preprocess(doc)
        for word in words:
            G[wordIndex[word]][docIndex[doc]] = 1

    return G, docIndex, wordIndex
```

```python
docs = ["./docs/libro" + str(i) + ".txt" for i in range(1,7)]
print(docs)

[G, docIndex, wordIndex] = buildMatrix(docs)
print(G)
```

## 4- Aplicar consultas booleanas

```python
def M(word):
    return G[wordIndex[word]]

def AND(A, B):
    res = np.zeros(len(docIndex), dtype=bool)
    for i in range(len(docIndex)):
        res = A[i] and B[i]
    return res

def OR(A, B):
    res = np.zeros(len(docIndex), dtype=bool)
    for i in range(len(docIndex)):
        res = A[i] or B[i]
    return res

def NOT(A):
    res = np.zeros(len(docIndex), dtype=bool)
    for i in range(len(docIndex)):
        res = not A[i]
    return res

Query1 = AND(M("anillo"), M("frodo"))
print("anillo AND frodo: ", Query1)
Query2 = OR(M("anillo"), M("frodo"))
print("anillo OR frodo: ", Query1)
# definir mas queries
```
